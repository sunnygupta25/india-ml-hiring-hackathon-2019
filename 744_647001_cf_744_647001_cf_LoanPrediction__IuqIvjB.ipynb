```python
#Importing important libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
pd.set_option('display.max_columns',100)
```

```python
#reading the train file
train = pd.read_csv('./train.csv')
train.head(2)
```

```python
#spliting categorical and number datatype columns

#train_num = train.select_dtypes(include='number')
num_cols = ['interest_rate','unpaid_principal_bal','loan_term','loan_to_value','number_of_borrowers','insurance_percent','co-borrower_credit_score','borrower_credit_score'
           ,'debt_to_income_ratio']
#if len(train.columns) == len(train_cat.columns)+len(train_num.columns):
#    print(True)
#else:
#    print(False)
train_num = train[num_cols]
train_num.head(2)
```

```python
def cap_outlier(array,k=3):
    upper_bond=array.mean()+k*array.std()
    lower_bond = array.mean()-k*array.std()
    array[array<lower_bond] = lower_bond
    array[array>upper_bond] = upper_bond
    return array

train_num = train_num.apply(cap_outlier,axis=0)
train_num.head(2)
```

```python
from sklearn.preprocessing import scale
train_num=scale(train_num)

train_num=pd.DataFrame(train_num,columns=num_cols)
train_num.head(2)
```

```python
cols = [col for col in train.columns if col not in num_cols]
cols
```

```python
train1 = train[cols]
```

```python

#train_num1 = pd.DataFrame(train_num,columns = num_cols)
#train_num1 = train_num1.apply(cap_outlier,axis=0)
#train_num1.head()
train2 = pd.concat([train1,train_num],axis=1)
train2.head(2)
```

```python
#train_cat = train2.select_dtypes(include='object')
train_cat=train2[['source','financial_institution','origination_date','first_payment_date','loan_purpose']]
train_cat.columns
#train2.info()
```

```python
print('Source:\n',train_cat.source.value_counts())
print('\nfinancial_institution:\n',train_cat.financial_institution.value_counts())
print('\norigination_date:\n',train_cat.origination_date.value_counts())
print('\nfirst_payment_date:\n', train_cat.first_payment_date.value_counts())
print('\nloan_purpose:\n', train_cat.loan_purpose.value_counts())
```

```python
source = pd.get_dummies(train2.source,prefix='source',drop_first=True)
source.head(2)
train2 = train2.drop('source',axis=1)

financial_institution = pd.get_dummies(train2.financial_institution,drop_first=True,prefix='financial_institution')
train2 = train2.drop('financial_institution',axis=1)
train2 = pd.concat([train2,source,financial_institution],axis=1)

origination_date = pd.get_dummies(train2.origination_date,prefix='origination_date',drop_first=True)
train2 = train2.drop('origination_date',axis=1)
train2=pd.concat([train2,origination_date],axis=1)

first_payment_date = pd.get_dummies(train2.first_payment_date,drop_first = True,prefix='first_payment_date')
train2 = train2.drop('first_payment_date',axis=1)
train2=pd.concat([train2,first_payment_date],axis=1)

loan_purpose = pd.get_dummies(train2.loan_purpose,prefix='loan_purpose',drop_first=True)
train2=train2.drop('loan_purpose',axis=1)
train2=pd.concat([train2,loan_purpose],axis=1)
train2.head(3)
```

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation, BatchNormalization
import keras
from keras.layers import Input, Dense
from keras.optimizers import SGD

model = Sequential()
model.add(Dense(8, input_dim=16))
model.add(Activation('relu'))
#model.add(BatchNormalization())
model.add(Dropout(0.2))
model.add(Dense(16))
model.add(Activation('relu'))
#model.add(BatchNormalization())
model.add(Dropout(0.2))
model.add(Dense(32))
model.add(Activation('relu'))
model.add(Dropout(0.2))
#model.add(Dense(64))
#model.add(Activation('relu'))
#model.add(Dropout(0.25))
#model.add(Dense(32))
#model.add(Activation('relu'))
#model.add(Dropout(0.25))
model.add(Dense(64))
model.add(Activation('softmax'))
model.add(BatchNormalization())
model.add(Dense(output_dim=1, init='normal', activation='sigmoid'))

#sgd = SGD(lr=0.01, decay=1e-4, momentum=0.7, nesterov=True)
#model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
#model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.fit(X_train_1,y_train,class_weight='balanced', epochs=200, batch_size=32, validation_data=(X_test_1, y_test))
```

```python
y_pred = model.predict(X_test_1)

y_pred = (y_pred > 0.2)

print('Classification Report:\n',classification_report(y_test,y_pred))
print('ConfusionMatrix:\n',confusion_matrix(y_test,y_pred))
```
